## Generating a line segment

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib notebook
```

```python
xs = []; ys = []; zs = []
for a in np.linspace(0, 1, 100):
    for b in np.linspace(0, 1, 100):
        for c in np.linspace(0, 1, 100):
            if abs(a + b + c - 1) < 0.0000000001:
                xs.append(a)
                ys.append(b)
                zs.append(c)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(xs, ys, zs)
ax.grid(True, which='both')
# ax.set_aspect('equal')
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')
# fig.savefig("line_segment.png")
plt.show()
```

```python
xs = []; ys = [];
for a in np.linspace(-1, 1, 2000):
    for b in np.linspace(-1, 1, 2000):
        if abs(a**2 + b**2 - .5) < .001:
            xs.append(a)
            ys.append(b)
xs = np.array(xs); ys = np.array(ys);
```

```python
fig, ax = plt.subplots()
plt.scatter(xs, ys)
ax.set_aspect('equal')
ax.set_xlim(-1.1, 1.1)
ax.set_ylim(-1.1, 1.1)
fig.set_size_inches(15, 15)
fig.savefig("deform_retract_circle2.png")
plt.show()
```
